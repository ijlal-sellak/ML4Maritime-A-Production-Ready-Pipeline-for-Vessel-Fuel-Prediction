# 🧾 Missing Values Handling — FuelCast Project

### 1️- Analyse du pourcentage de valeurs manquantes

Description :
Cette étape permet de mesurer la proportion de valeurs manquantes pour chaque variable afin d’identifier les capteurs ou mesures les plus défaillants.

```python
# Calculate missing percentage per column
missing_percent = (
    df_train.isnull().mean() * 100
).sort_values(ascending=False)

# Display top 20 columns with highest missing values
missing_percent.head(20)
```

### 2️ -Visualisation des colonnes les plus impactées

Description :
Une visualisation facilite l’identification rapide des variables nécessitant un traitement spécifique.

```python
# Select top missing columns
top_missing = missing_percent.head(15)

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_missing.values,
    y=top_missing.index
)

plt.title("Top Columns with Missing Values")
plt.xlabel("Missing Percentage (%)")
plt.ylabel("Columns")

plt.show()
```

### 3-Segmentation des colonnes selon le taux de Missing

Description :
Les colonnes sont classées en trois catégories pour appliquer des stratégies d’imputation adaptées.

```python
# High missing > 15%
high_missing = missing_percent[
    missing_percent > 15
].index.tolist()

# Medium missing 5–15%
medium_missing = missing_percent[
    (missing_percent > 5) &
    (missing_percent <= 15)
].index.tolist()

# Low missing < 5%
low_missing = missing_percent[
    missing_percent <= 5
].index.tolist()

print("High Missing Columns:", len(high_missing))
print("Medium Missing Columns:", len(medium_missing))
print("Low Missing Columns:", len(low_missing))
```

### 4️- Suppression des colonnes très vides (optionnel)

Description :
Si certaines variables dépassent 40 % de valeurs manquantes, elles peuvent être supprimées car peu exploitables.

```python
threshold = 0.40

df_train = df_train.loc[
    :,
    df_train.isnull().mean() < threshold
]
```

### 5️- Identification des colonnes numériques

Description :
Les méthodes d’imputation statistiques s’appliquent uniquement aux variables numériques.

```python
num_cols = df_train.select_dtypes(
    include="number"
).columns

print("Number of numerical columns:", len(num_cols))
```

### 6️⃣ Imputation simple — Colonnes à faible missing

Méthode : Médiane

Description :
La médiane est robuste face aux valeurs extrêmes fréquentes dans les données mécaniques.

```python
for col in low_missing:
    if col in num_cols:
        df_train[col].fillna(
            df_train[col].median(),
            inplace=True
        )
```

7- Imputation avancée — KNN Imputer

Description :
Les variables moteurs et propulsion étant corrélées, KNN permet une estimation plus réaliste des valeurs manquantes.

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)

# Get numerical columns from the current df_train state for imputation
current_num_cols = df_train.select_dtypes(include="number").columns

df_train[current_num_cols] = imputer.fit_transform(
    df_train[current_num_cols]
)

# After imputation, check for remaining missing values to ensure no NaNs are left
print("Missing values after KNN imputation:", df_train[current_num_cols].isnull().sum().sum())
```

### 8️- Interpolation — Variables de navigation

Description :
Les données de trajectoire du navire évoluent de manière continue, d’où l’utilisation d’une interpolation linéaire.

```python
nav_cols = [
    col for col in df_train.columns
    if "Ship_" in col
]

df_train[nav_cols] = df_train[
    nav_cols
].interpolate(method="linear")
```

### 9️- Interpolation — Variables météorologiques

Description :
Les conditions météo varient progressivement dans le temps, ce qui justifie l’interpolation.

```python
weather_cols = [
    col for col in df_train.columns
    if "Weather_" in col
]

df_train[weather_cols] = df_train[
    weather_cols
].interpolate(method="linear")
```

### - Vérification finale des valeurs manquantes

Description :
Cette étape confirme que toutes les valeurs manquantes ont été correctement traitées.

```python
# Total remaining missing values
total_missing = df_train.isnull().sum().sum()

print("Remaining Missing Values:", total_missing)
```

## Outliers Detection

### 1️- Détection des outliers — Méthode IQR

Description :
La méthode IQR (Interquartile Range) permet d’identifier les valeurs extrêmes situées en dehors de l’intervalle statistique normal.

Formule :

IQR = Q3 − Q1

Outlier < Q1 − 1.5×IQR

Outlier > Q3 + 1.5×IQR

```python
# Select numerical columns
num_cols = df_train.select_dtypes(
    include="number"
).columns

# Detect outliers count per column
outliers_count = {}

for col in num_cols:

    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_train[
        (df_train[col] < lower_bound) |
        (df_train[col] > upper_bound)
    ]

    outliers_count[col] = outliers.shape[0]

# Display columns with most outliers
pd.Series(outliers_count).sort_values(
    ascending=False
).head(10)
```

### 4-Définition de la fonction Winsorization

Description :
Remplace les valeurs extrêmes par les percentiles 1 % et 99 %.

```python
def winsorize_column(series, lower_q=0.01, upper_q=0.99):

    lower = series.quantile(lower_q)
    upper = series.quantile(upper_q)

    return series.clip(lower, upper)
```

### 5- Application du capping

Description :
On applique la Winsorization à toutes les variables numériques.

```python
for col in num_cols:

    df_train[col] = winsorize_column(
        df_train[col]
    )
```

## Feature Engineering Pipeline

### 1️- Objectif du Feature Engineering

Description :
Le Feature Engineering consiste à créer de nouvelles variables plus explicatives à partir des données brutes afin d’améliorer la performance des modèles de prédiction de consommation fuel.

Dans le contexte FuelCast, nous combinons :

Charges moteurs

Puissance propulsion

Conditions météo

Dynamique du navire

### 2️- Total Engine Fuel Consumption

Description :
On agrège la consommation instantanée de tous les générateurs pour obtenir une mesure globale de la charge énergétique.

```python
engine_fuel_cols = [
    col for col in df_train.columns
    if "MomentaryFuel" in col and "GeneratorEngine" in col
]

df_train["Total_Engine_Fuel"] = df_train[
    engine_fuel_cols
].sum(axis=1)

print("Engine fuel columns used:", engine_fuel_cols)
```

### 3️- Total Shaft Power — Engines

Description :
Représente la puissance mécanique totale fournie par les moteurs.

```python
engine_power_cols = [
    col for col in df_train.columns
    if "ShaftPower" in col and "GeneratorEngine" in col
]

df_train["Total_Engine_ShaftPower"] = df_train[
    engine_power_cols
].sum(axis=1)
```

### 4️ Total Propeller Shaft Power

Description :
Puissance totale fournie aux hélices (propulsion réelle).

```python
propeller_power_cols = [
    "Propeller_Port_ShaftPower",
    "Propeller_Starboard_ShaftPower"
]

df_train["Total_Propeller_Power"] = df_train[
    propeller_power_cols
].sum(axis=1)
```

### 5️- Propulsion Balance Ratio

Description :
Mesure l’équilibre entre hélice bâbord et tribord.

Déséquilibre → surconsommation fuel.

```python
df_train["Propulsion_Balance_Ratio"] = (
    df_train["Propeller_Port_ShaftPower"] /
    (df_train["Propeller_Starboard_ShaftPower"] + 1e-6)
)
```

### 6️-Weather Severity Index

Description :
Indice combinant houle et vent pour mesurer la sévérité des conditions maritimes.

```python
df_train["Weather_Severity_Index"] = (
    df_train["Weather_WaveHeight"] *
    df_train["Weather_WindSpeed10M"]
)
```

### 7️- Ocean Resistance Index

Description :
Mesure la résistance hydrodynamique due aux courants.

```python
df_train["Ocean_Resistance_Index"] = (
    df_train["Weather_OceanCurrentVelocity"] *
    df_train["Ship_SpeedOverGround"]
)
```

### 8️- Effective Propulsion Efficiency

Description :
Compare la puissance propulsion à la vitesse réelle du navire.

```python
df_train["Propulsion_Efficiency"] = (
    df_train["Ship_SpeedOverGround"] /
    (df_train["Total_Propeller_Power"] + 1e-6)
)
```

### 9️- Liste des nouvelles colonnes

```python
new_features = [
    "Total_Engine_Fuel",
    "Total_Engine_ShaftPower",
    "Total_Propeller_Power",
    "Propulsion_Balance_Ratio",
    "Weather_Severity_Index",
    "Ocean_Resistance_Index",
    "Propulsion_Efficiency"
]

df_train[new_features].head()
```

Description :
Les variables issues du Feature Engineering présentent des distributions très asymétriques (skewed), avec des valeurs extrêmes élevées.

La transformation permet de :

Réduire la skewness

Stabiliser la variance

Améliorer la performance des modèles linéaires

### 2- Log Transformation — Balance Ratio

Description :
On applique une transformation logarithmique pour réduire l’asymétrie de la variable représentant l’équilibre de propulsion.

```python
df_train["Propulsion_Balance_Ratio_log"] = np.log1p(
    df_train["Propulsion_Balance_Ratio"]
)
```

### 3- Power Transform — Propulsion Efficiency

Description :
La transformation logarithmique étant insuffisante pour cette variable très asymétrique, on applique une transformation de puissance (Yeo-Johnson).

Cette méthode :

Gère les zéros

Réduit les extrêmes

Normalise la distribution

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer()

df_train["Propulsion_Efficiency_pt"] = pt.fit_transform(
    df_train[["Propulsion_Efficiency"]]
)
```

### Vérification après Power Transform

```python
df_train["Propulsion_Efficiency_pt"] = df_train[
    "Propulsion_Efficiency_pt"
].clip(-3, 3)

```

**Groupe moteurs – Fuel & Power**

Les variables liées aux moteurs, notamment Consumer_GeneratorEngineX_MomentaryFuel, Consumer_GeneratorEngineX_ShaftPower, Consumer_Total_ShaftPower et Total_Engine_Fuel, présentent des corrélations très fortes, comprises entre 0.7 et 0.95. Cette relation est parfaitement cohérente d’un point de vue physique : lorsqu’un moteur fournit davantage de puissance mécanique, il consomme mécaniquement plus de carburant. De plus, les moteurs fonctionnent généralement de manière simultanée, ce qui entraîne une forte colinéarité entre les puissances et consommations individuelles et leurs équivalents agrégés. D’un point de vue data science, conserver les variables globales (Total_Engine_Fuel et Consumer_Total_ShaftPower) est suffisant, tandis que les mesures individuelles par moteur sont redondantes et peuvent être supprimées sans perte d’information.

**Groupe propulsion – Hélices et transmission**

Les variables décrivant la propulsion, telles que Propeller_Port_ShaftPower, Propeller_Starboard_ShaftPower, Propeller_Total_ShaftPower, ainsi que les couples (ShaftTorque) et vitesses de rotation (RotationSpeed), sont fortement corrélées entre elles, avec des coefficients dépassant souvent 0.85. Cette corrélation s’explique par la relation mécanique directe entre couple, vitesse de rotation et puissance, ainsi que par le fonctionnement synchronisé des hélices bâbord et tribord. D’un point de vue analytique, la variable de puissance totale des hélices (Total_Propeller_Power) synthétise efficacement l’information utile. Les variables détaillées par hélice ou par composante mécanique sont donc supprimées afin d’éviter la redondance et la multicolinéarité.

**Navigation – Vitesse du navire**

Les variables Ship_SpeedThroughWater et Ship_SpeedOverGround présentent une corrélation quasi parfaite, proche de 0.9. Cette similarité s’explique par le fait que les courants marins sont généralement faibles ou stables sur la majorité des observations, rendant les deux mesures très proches. Dans une logique de simplification et de robustesse du modèle, seule la vitesse par rapport au sol (Ship_SpeedOverGround), plus globale et directement exploitable, est conservée, tandis que la vitesse par rapport à l’eau est supprimée.

**Groupe météo – Vagues et vent**

Les variables météorologiques liées à l’état de la mer, telles que les hauteurs et périodes de vagues (WaveHeight, WavePeriod, SwellWaveHeight, WindWaveHeight, etc.), montrent de fortes corrélations au sein de chaque sous-groupe. Leur corrélation avec la consommation et la propulsion est modérée mais significative, traduisant leur rôle indirect dans l’augmentation de la résistance hydrodynamique du navire. Afin de conserver une représentation pertinente sans redondance excessive, seules les variables les plus explicites et globales (Weather_WaveHeight, Weather_WavePeriod et Weather_WindSpeed10M) sont retenues, tandis que les mesures redondantes de type swell ou wind wave sont supprimées.

**Groupe météo – Radiation solaire**

Les variables Weather_DirectNormalIrradiance et Weather_ShortwaveRadiation sont extrêmement corrélées, avec un coefficient supérieur à 0.9, car elles décrivent le même phénomène énergétique solaire sous des formes légèrement différentes. Dans ce contexte, conserver une seule variable est suffisant. La radiation solaire globale (Weather_ShortwaveRadiation) est privilégiée pour sa portée plus générale, tandis que la radiation directe normale est supprimée.

**Features engineered – Validation finale**

Les variables issues du feature engineering, telles que Total_Engine_Fuel, Total_Propeller_Power, Weather_Severity_Index et Propulsion_Efficiency_pt, présentent des corrélations cohérentes avec la consommation de carburant et la puissance, sans atteindre de corrélations parfaites supérieures à 0.95 entre elles. Elles constituent ainsi un signal prédictif de grande qualité et sont conservées intégralement pour la phase de modélisation.

Ajustements Non Linéaires — Comparaison des Corrélations

Description :
Pour choisir le meilleur type d'ajustement, on compare les coefficients de corrélation linéaire après transformation des variables :

- Linéaire    : r(x, y)
- Exponentiel : r(x, ln y)
- Puissance   : r(ln x, ln y)
- Logarithmique : r(ln x, y)

```python
x = df_train_reduced["Ship_SpeedOverGround"].replace(0, np.nan).dropna()
y_target = df_train_reduced.loc[x.index, "Consumer_Total_MomentaryFuel"].replace(0, np.nan).dropna()

# Aligner les deux séries
common_idx = x.index.intersection(y_target.index)
x_clean = x.loc[common_idx]
y_clean = y_target.loc[common_idx]

r_linear  = np.corrcoef(x_clean, y_clean)[0, 1]
r_exp     = np.corrcoef(x_clean, np.log(y_clean))[0, 1]
r_power   = np.corrcoef(np.log(x_clean), np.log(y_clean))[0, 1]
r_log     = np.corrcoef(np.log(x_clean), y_clean)[0, 1]

print("=== Comparaison des ajustements ===")
print(f"Linéaire      r(x, y)         : {r_linear:.4f}")
print(f"Exponentiel   r(x, ln y)      : {r_exp:.4f}")
print(f"Puissance     r(ln x, ln y)   : {r_power:.4f}")
print(f"Logarithmique r(ln x, y)      : {r_log:.4f}")

best = max(
    [("Linéaire", abs(r_linear)), ("Exponentiel", abs(r_exp)),
     ("Puissance", abs(r_power)), ("Logarithmique", abs(r_log))],
    key=lambda t: t[1]
)
print(f"\nMeilleur ajustement : {best[0]} (|r| = {best[1]:.4f})")
```

## Partie 3 — Régression Polynomiale

Description :
On implémente la régression polynomiale (degrés 1, 2, 3, 4) sur l'ensemble du dataset (X_train / X_test).  
On utilise `PolynomialFeatures` de scikit-learn combiné à un pipeline.

Visualisation — Courbes de régression polynomiale (degrés 1 à 4)

Description :
On entraîne un modèle polynomial sur la variable `Ship_SpeedOverGround` seule pour visualiser les courbes d'ajustement pour différents degrés.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import numpy as np

x_plot = x_clean.values.reshape(-1, 1)
y_plot = y_clean.values

x_range = np.linspace(x_clean.min(), x_clean.max(), 300).reshape(-1, 1)

plt.figure(figsize=(12, 5))
plt.scatter(x_plot, y_plot, alpha=0.2, label="Données", color="gray", s=5)

colors = ["blue", "green", "orange", "red"]

for degree, color in zip([1, 2, 3, 4], colors):
    poly_pipeline = Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("lr", LinearRegression())
    ])
    poly_pipeline.fit(x_plot, y_plot)
    y_range_pred = poly_pipeline.predict(x_range)

    plt.plot(x_range, y_range_pred, label=f"Degré {degree}", color=color, linewidth=2)

plt.title("Régression Polynomiale — Vitesse vs Consommation Fuel")
plt.xlabel("Ship_SpeedOverGround")
plt.ylabel("Consumer_Total_MomentaryFuel")
plt.legend()
plt.tight_layout()
plt.show()
```

Comparaison des modèles polynomiaux — R² et RMSE

Description :
On entraîne un modèle polynomial de degré 1 à 4 sur **toutes les features** (X_train / X_test) et on compare les métriques R² et RMSE.

```python
from sklearn.metrics import r2_score, mean_squared_error

results = []

for degree in [1, 2, 3, 4]:
    poly_pipe = Pipeline([
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("scaler", StandardScaler()),
        ("lr", LinearRegression())
    ])
    poly_pipe.fit(X_train, y_train)
    y_pred_poly = poly_pipe.predict(X_test)

    r2   = r2_score(y_test, y_pred_poly)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred_poly))

    results.append({
        "Degré": degree,
        "R²": round(r2, 4),
        "RMSE": round(rmse, 4)
    })
    print(f"Degré {degree} → R² = {r2:.4f} | RMSE = {rmse:.4f}")

import pandas as pd
results_df = pd.DataFrame(results)
results_df
```

Application du Meilleur Modèle

Description :
D'après la comparaison des métriques, on sélectionne le modèle polynomial au meilleur degré (celui avec le R² le plus élevé sans surapprentissage) et on l'applique comme modèle final.

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Choisir le meilleur degré basé sur les résultats ci-dessus
best_degree = results_df.loc[results_df["R²"].idxmax(), "Degré"]
print(f"Meilleur degré sélectionné : {best_degree}")

best_poly_pipeline = Pipeline([
    ("poly", PolynomialFeatures(degree=best_degree, include_bias=False)),
    ("scaler", StandardScaler()),
    ("lr", LinearRegression())
])

best_poly_pipeline.fit(X_train, y_train)
y_pred_best = best_poly_pipeline.predict(X_test)

r2_best   = r2_score(y_test, y_pred_best)
rmse_best = np.sqrt(mean_squared_error(y_test, y_pred_best))
mae_best  = mean_absolute_error(y_test, y_pred_best)

print(f"\n=== Modèle Final — Polynomial Degré {best_degree} ===")
print(f"R²   : {r2_best:.4f}")
print(f"RMSE : {rmse_best:.4f}")
print(f"MAE  : {mae_best:.4f}")
```

Description :
Récapitulatif final comparant la régression linéaire et les modèles polynomiaux sur les métriques R², RMSE et MAE.
